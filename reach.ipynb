```python
# 环境测试
from reachEnv import *
from robosuite.controllers import load_controller_config
controller_config = load_controller_config(default_controller="OSC_POSE")

env = ReachEnv(controller_config=controller_config,seed=2)

env._test_matrixs()
obs = env.reset()
action = np.zeros(env.action_dim)
obs, reward, done, _ = env.step(action)

joint_positions = env.robots[0].controller.joint_pos
joint_velocities = env.robots[0].controller.joint_vel
print(joint_positions)
print(joint_velocities)
for i in range(1000):
    action = np.zeros(env.action_dim)
    obs, reward, done, _ = env.step(action)
    if i % 20 == 0:
        print("reward:", reward)
        print("done:", done)
    if done:
        break

    env.render()

env.close()
```
